```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get('https://picklebet.com/sports/mma/betting?tab=next&page=1')
driver.implicitly_wait(10)

links = [i.get_attribute('href') for i in driver.find_elements_by_xpath("//div[@class='MatchRow-module--countdown--NQwKx']/a")]

for a in links:
    driver.get(a)
    for i in driver.find_elements_by_class_name('MarketGroup-module--group--2TkGm'):
        i.click()

    others = []
    for i in range(len(driver.find_elements_by_class_name('MarketGroup-module--group--2TkGm'))):
        lst = driver.find_elements_by_class_name('MarketGroup-module--group--2TkGm')[i].text.split('\n')
        if i == 0:
            Player1, Win1, Player2, Win2 = lst[3], lst[4], lst[-2], lst[-1]
            continue
        for j in lst:
            if j == '1' or j == 'Total' or j == 'Additional Markets':
                continue
            others.append(j)

    print(f"{Player1} vs {Player2} | Win: {Win1}/ Win: {Win2}, {others}")
    print()
    
driver.get('https://picklebet.com/sports/mma/betting?tab=next&page=1')
for ind, i in enumerate(driver.find_elements_by_class_name('MatchRow-module--match--zMgI3')):
    if '+' not in i.text:   
        match = driver.find_elements_by_class_name('Outcomes-module--outcomes--2bGtG')[ind].text.split('\n')
        Player1, Win1, Player2, Win2 = match[0], match[1], match[-2], match[-1]

        print(f"{Player1} vs {Player2} | Win: {Win1}/ Win: {Win2}")
        print()
```
