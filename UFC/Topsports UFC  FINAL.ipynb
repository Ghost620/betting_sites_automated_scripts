```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import time
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.maximize_window()

lst = ['https://www.topsport.com.au/Sport/Mixed_Martial_Arts/UFC/Matches', 'https://www.topsport.com.au/Sport/Mixed_Martial_Arts/UFC_269_-_Main_Card_/Matches', 'https://www.topsport.com.au/Sport/Mixed_Martial_Arts/UFC_270_-_Main_Card/Matches', 'https://www.topsport.com.au/Sport/Mixed_Martial_Arts/UFC_Fight_Night_198_-_Main_Card/Matches', 'https://www.topsport.com.au/Sport/Mixed_Martial_Arts/UFC_Fight_Night_198_-_Preliminary_Card/Matches', 'https://www.topsport.com.au/Sport/Mixed_Martial_Arts/UFC_on_ESPN_31_-_Main_Card/Matches', 'https://www.topsport.com.au/Sport/Mixed_Martial_Arts/UFC_on_ESPN_31_-_Preliminary_Card/Matches']    
def Add_Mar(x):
    driver.find_elements_by_class_name('additMark')[x].click()
    time.sleep(5)

    for i in driver.find_elements_by_class_name('icon-navToggle'):
        if 'Total Rounds' in i.text:
            i.click()
        if 'Round Betting' in i.text:
            i.click()
        if 'Double Chance' in i.text:
            i.click()
        if 'Method of Victory' in i.text:
            i.click()

    time.sleep(1)

    for j in driver.find_elements_by_tag_name('tbody'):
        try:
            if len(j.text) > 100:
                others.append(j.text.split('\n'))
            if 'Rounds' in j.text:
                others.append(j.text.split('\n'))
        except:
            continue

    time.sleep(1)

    return others

for k in lst:
    driver.get(k)
    driver.implicitly_wait(15)
    
    league_name = driver.find_element_by_xpath("//h1[@class='icon icon-MMA']").text
    print(f"____________________________________ {league_name} ________________________________________")
    print("")
    a = 0
    try:
        for i in range(len(driver.find_elements_by_class_name('match'))):
            try:
                match = driver.find_elements_by_class_name('match')[i].text.split('\n')
                Player1, Win1, Player2, Win2 = match[1], match[2], match[3], match[4]
                
                print(f"BET{i+1} ---> Player1: {Player1}, Win: {Win1}, Player2: {Player2}, Win: {Win2}")
                
                if 'Additional Market' in match[-1]:
                    others = []
                    Add_Mar(a)
                    a+=1
                    print(others)

            except:
                a+=1
                continue
        print("-------------------------------------------------------------------------------------------------------")
    except:
        print('URL Denied')
```
