```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import time
driver = webdriver.Chrome()
# leagues = ['England/English-Premier-League/56', 'Germany/Germany-1.-Bundesliga/96', 'Germany/2.Bundesliga/97', 'Spain/La-Liga-Santander/61', 'Italy/Serie-A-TIM/82', 'Europe/UEFA-Champions-League/86', 'Europe/UEFA-Europa-League/94', 'Brazil/Série-A/100', 'Austria/Tipico-Bundesliga/373', 'Norway/Eliteserien/3185', 'Sweden/Superettan/2125', 'Sweden/Allsvenskan/2124', 'Netherlands/Eredivisie/394']

leagues = ['https://www.betright.com.au/sports/Soccer/England/English-Premier-League/56', 'https://www.betright.com.au/sports/Soccer/England/EFL-Trophy/108', 'https://www.betright.com.au/sports/Soccer/England/Sky-Bet-League-1/58', 'https://www.betright.com.au/sports/Soccer/England/Sky-Bet-League-2/95']
for i in leagues:
    driver.get(f'https://www.betright.com.au/sports/Soccer/{i}')
    driver.implicitly_wait(10)
    time.sleep(10)
    name = driver.find_element_by_xpath("//h2").text
    print("---------------------------- {name} -----------------------------------")
    print()
    for k in range(len(driver.find_elements_by_xpath("//div[@class='ctr--epsilon  soft']"))):
        match = driver.find_elements_by_xpath("//section[@class='push--bottom ng-scope']")[k].text.split('\n')
        Team1, Win1, Draw, Team2, Win2 = match[5], match[6], match[8], match[9], match[10]
        driver.find_elements_by_xpath("//div[@class='ctr--epsilon  soft']")[k].click()
        time.sleep(10)
        for j in range(len(driver.find_elements_by_xpath("//li[@class='flex-matrix__item ng-scope']"))):
            if 'Totals' in driver.find_elements_by_xpath("//li[@class='flex-matrix__item ng-scope']")[j].text:
                driver.find_elements_by_xpath("//li[@class='flex-matrix__item ng-scope']/a")[j].click()
                time.sleep(10)
                lst = []
                try:
                    for i in range(70, 140):
                        if driver.find_elements_by_xpath("//div[@class='ng-scope']")[i].text.split('\n')[0] == 'Over/Under' or driver.find_elements_by_xpath("//div[@class='ng-scope']")[i].text.split('\n')[0] == 'Total Goals Over/Under':
                            lsts = driver.find_elements_by_xpath("//div[@class='ng-scope']")[i].text.split('\n')
                            del lsts[0]
                            del lsts[0]
                            if lsts not in lst:
                                lst.append(lsts)
                except:
                    continue

        driver.back()
        time.sleep(10)
        driver.back()
        time.sleep(10)
        print(f"BET{k+1} ---> {Team1} vs {Team2} | Win: {Win1}/ Draws: {Draw}/ Win: {Win2} & {lst}")
        print()
    print('--------------------------------------------------------------------------------------------------------------')
```
