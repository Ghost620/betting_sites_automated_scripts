```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import time
driver = webdriver.Chrome()
driver.get('https://picklebet.com/sports/soccer/betting/?tour=eflcup-21-22,premier-league-21-22&tab=next&page=5')
driver.implicitly_wait(30)
time.sleep(3)
for i in range(len(driver.find_elements_by_class_name('MatchRow-module--matchLink--2XbrR'))):
    driver.find_elements_by_class_name('MatchRow-module--matchLink--2XbrR')[i].click()
    
    match = driver.find_element_by_class_name('MarketGroup-module--outcomes--3ye0g').text.split('\n')
    Team1, Win1, Draw, Team2, Win2 = match[0], match[1], match[3], match[4], match[5]
    
    Over_Under = []
    for j in driver.find_elements_by_class_name('MarketGroup-module--group--2TkGm'):
        if 'Total' in j.text:
            j.click()
            for k in j.text.split('\n'):
                if 'Total' in k or k == '5':
                    continue
                Over_Under.append(k)
        
    print(f"BET{i+1} ---> {Team1} vs {Team2} | Win: {Win1}/ Draws: {Draw}/ Win: {Win2} & {Over_Under}")
    print()
    driver.back()
```
