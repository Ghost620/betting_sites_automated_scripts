```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import time
driver = webdriver.Chrome()
#league = ['Brazilian_Serie_A', 'Dutch_Eerste_Divisie', 'English_League_Championship', 'English_League_One', 'English_Premier_League', 'French_Ligue_1', 'French_Ligue_2', 'German_Bundesliga', 'Italian_Serie_A', 'Italian_Serie_B', 'Japanese_J1_League', 'UEFA_Champions_League', 'UEFA_Europa_League', 'Swedish_Allsvenskan', 'Swedish_Superettan', 'Portuguese_Primeira_Liga']

league = ['https://www.playup.com.au/Sport/Soccer/English_Premier_League/Matches', 'https://www.playup.com.au/Sport/Soccer/English_League_One/Matches', 'https://www.playup.com.au/Sport/Soccer/English_League_Two/Matches', 'https://www.playup.com.au/Sport/Soccer/English_EFL_Trophy/Matches']
for a in league:
    driver.get(f"https://www.playup.com.au/Sport/Soccer/{a}/Matches")
    driver.implicitly_wait(15)
    print(f'-------------------------- {a} --------------------------')
    print()

    for j in range(len(driver.find_elements_by_xpath("//table[@class='MarketTable MatchMarket']"))):
        try:
            lst = driver.find_elements_by_xpath("//table[@class='MarketTable MatchMarket']")[j].text.split('\n')
            Team1, Win1, Draw, Team2, Win2 = lst[1], lst[2], lst[3], lst[6], lst[7]
            Over, Under = lst[5], lst[-1]
        except:
            try:
                lst = driver.find_elements_by_xpath("//table[@class='MarketTable MatchMarket']")[12].text.split('\n')
                Team1, Win1, Draw, Team2, Win2 = lst[1], lst[2], lst[3], lst[5], lst[6]
                Over, Under = lst[4], lst[-1]
            except:
                lst = driver.find_elements_by_xpath("//table[@class='MarketTable MatchMarket']")[j].text.split('\n')
                Team1, Win1, Draw, Team2, Win2 = lst[1], lst[2], lst[3], lst[4], lst[5]
                Over, Under = '-', '-'

        print(f"BET{j+1} ---> {Team1} vs {Team2} | Win: {Win1}/ Draws: {Draw}/ Win: {Win2} & {Over}/ {Under}")
    print()
```
