```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.maximize_window()
driver.get('https://picklebet.com/sports/golf/betting?tab=next&page=1')

driver.implicitly_wait(30)

lst = [i.get_attribute('href') for i in driver.find_elements_by_xpath("//*[@id='gatsby-focus-wrapper']/div/div[1]/div[1]/div/div/div/div/div[2]/div[4]/a")]
for i in lst:
    Players, Top5, Top10, Top20, Winner, groups, t10, t20, t5 = ['Players'], ['Top5'], ['Top10'], ['Top20'], ['Winner'], [], [], [], []
    driver.get(i)
    name = driver.find_element_by_tag_name('h1').text
    print(f'--------------------------- {name} ------------------------------')
    print()
    
    try:
        sections = driver.find_elements_by_class_name('MarketGroup-module--group--2TkGm')
        if len(sections) == 1:
            winn_lst = sections[0].text.split('\n')
            for k in range(3, len(winn_lst)):
                if k % 2:
                    Players.append(winn_lst[k])
                else:
                    Winner.append(winn_lst[k])
            for j in range(len(Players)):
                print(f"{Players[j]} |  {Winner[j]}")
            print()
            
        for j in driver.find_elements_by_class_name('MarketGroup-module--group--2TkGm'):
            if 'Top' in j.text or '72 Hole Group' in j.text or '72 Hole Match' in j.text: 
                j.click()

        for j in driver.find_elements_by_class_name('MarketGroup-module--group--2TkGm'):
            if 'Top 10' in j.text: 
                t10 = j.text.split('\n')
            if 'Top 20' in j.text: 
                t20 = j.text.split('\n')
            if 'Top 5' in j.text: 
                t5 = j.text.split('\n')
            if '72 Hole Group' in j.text:
                for a in j.text.split('\n'):
                    if len(a) <= 2:
                        continue
                    groups.append(a)

            if '72 Hole Match' in j.text:
                match = j.text.split('\n')
                Player1, Win1, Draws, Player2, Win2 = match[3], match[4], match[-1], match[5], match[6]
                print(Player1, Win1, Draws, Player2, Win2)
                print()

        for k in range(3, len(t10)):
            if k % 2:
                Players.append(t10[k])
            else:
                if len(t5) != 0:
                    Top5.append(t5[k])
                else:
                    Top5.append('-')
                if len(t10) != 0:
                    Top10.append(t10[k])
                else:
                    Top10.append('-')
                if len(t20) != 0:
                    Top20.append(t20[k])
                else:
                    Top20.append('-')

        for j in range(len(Players)):
            print(f"{Players[j]} |  {Top5[j]}  | {Top10[j]}  | {Top20[j]}")
        print()
        print(groups)
    except:
        continue
```
