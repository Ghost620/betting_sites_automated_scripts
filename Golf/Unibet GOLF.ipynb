```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import time
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get('https://www.unibet.com.au/betting/sports/filter/golf/all/competitions')
driver.implicitly_wait(5)

driver.find_element_by_id('CybotCookiebotDialogBodyLevelButtonLevelOptinAllowAll').click()

def view():
    for i in range(len(driver.find_elements_by_xpath("//div[@class='c1d1b _8651b']"))):
        try:
            driver.find_elements_by_xpath("//div[@class='c1d1b _8651b']")[i].click()
        except:
            continue


view()
for j in range(len(driver.find_elements_by_xpath("//*[text()='View all']"))):
    if j == 0:
        pass
    else:
        view()
        
    element = driver.find_elements_by_xpath("//*[text()='View all']")[j]
    actions = ActionChains(driver)
    actions.move_to_element(element).perform()
    element.click()
    
    try:
        
        driver.find_element_by_xpath("//div[@class='KambiBC-react-collapsable-container__show-more KambiBC-react-collapsable-container__show-more-outrights']").click()

        Players = driver.find_elements_by_class_name('KambiBC-outcomes-list__column')[0].text.split('\n')
        Winners = driver.find_elements_by_class_name('KambiBC-outcomes-list__column')[1].text.split('\n')
        Top5 = driver.find_elements_by_class_name('KambiBC-outcomes-list__column')[2].text.split('\n')
        Top10 = driver.find_elements_by_class_name('KambiBC-outcomes-list__column')[3].text.split('\n')
        Top20 = driver.find_elements_by_class_name('KambiBC-outcomes-list__column')[4].text.split('\n')
        del Winners[0], Top5[0], Top10[0], Top20[0]
        
        print(driver.find_element_by_class_name('_9f33f').text)
        print(Players, Top5, Top10, Top20)
    except:
        pass
               
    driver.back()

```
