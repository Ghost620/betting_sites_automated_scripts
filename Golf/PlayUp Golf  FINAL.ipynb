```python
# from selenium import webdriver
# from webdriver_manager.chrome import ChromeDriverManager
# from selenium.webdriver.common.keys import Keys
# import time, re 
# driver = webdriver.Chrome(ChromeDriverManager().install())
# driver.get('https://www.playup.com.au/Sport/Golf/')
# driver.implicitly_wait(30)

links = []
lst = [i.get_attribute('onclick') for i in driver.find_elements_by_tag_name('td')]
for j in lst:
    if j == None:
        continue
    links.append(re.findall(r"\'(.+?)\'", j)[0])

for a in links:
    driver.get(f"https://www.playup.com.au{a}")
    driver.implicitly_wait(30)
    
    name = driver.find_elements_by_tag_name('h1')[2].text
    print(f'------------------------------ {name} -------------------------------')
    print()
    try:   
        if 'MATCH' in name:
            for i in range(len(driver.find_elements_by_class_name('match'))):
                lst = driver.find_elements_by_class_name('match')[i].text.split('\n')
                Player1, Win1, Player2, Win2 = lst[-4], lst[-3], lst[-2], lst[-1]
                print(f"BET{i+1} ---> {Player1} vs {Player2}  |  {Win1}/ {Win2}")

        else:
            Players, Odds = [], []
            for j in range(len(driver.find_elements_by_tag_name('tr'))):
                if driver.find_elements_by_tag_name('tr')[j].text == '' or 'Win' in driver.find_elements_by_tag_name('tr')[j].text:
                    continue
                Players.append(driver.find_elements_by_tag_name('tr')[j].text.split('\n')[0])
                Odds.append(driver.find_elements_by_tag_name('tr')[j].text.split('\n')[1])

            for a in range(len(Players)):
                print(f"{Players[a]} | {Odds[a]}")

    except:
        try:
            driver.find_element_by_class_name('majorAction').click()
            time.sleep(4)
            for i in driver.find_elements_by_class_name('static'):
                if 'Top' in i.text or 'Winner' in i.text:
                    i.click()

            time.sleep(4)
            for ind, i in enumerate(driver.find_elements_by_tag_name('table')):
                if i.text == '' or "More Markets" in i.text:
                    continue
                elif ind == 2:
                    t10 = i.text.split('\n')
                elif ind == 3:
                    t20 = i.text.split('\n')
                elif ind == 4:
                    t5 = i.text.split('\n')
                elif ind == 5:
                    winn = i.text.split('\n')

            Players, Winner, Top5, Top10, Top20 = ['Players'], ['Winner'], ['TOP5'], ['TOP10'], ['TOP20']
            for i in range(len(winn)):
                if i == 0 or i == 1:
                    continue
                if i % 2 == 0:
                    Players.append(winn[i])
                else:
                    Top5.append(t5[i])
                    Top10.append(t10[i])
                    Top20.append(t20[i])
                    Winner.append(winn[i])

            for k in range(len(Players)):
                print(f"{Players[k]}  |  {Winner[k]}  |  {Top5[k]}  |  {Top10[k]}  |  {Top20[k]}")

        except:
            continue
```
