```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import time
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get('https://www.tab.com.au/sports/betting/Golf/competitions')
driver.implicitly_wait(100)
time.sleep(5)

for i in range(len(driver.find_elements_by_class_name('tbc-ui-details '))):
    driver.find_elements_by_class_name('tbc-ui-details ')[i].click()
    time.sleep(3)
    
    name = driver.find_element_by_class_name('competition-name').text
    print(f'------------------------------- {name} ------------------------------------')
    print()
    
    Players, Wins, Top5, Top10, Top20 = ['Players'], ['Wins'], ['Top5'], ['Top10'], ['Top20']
    if driver.find_elements_by_class_name('proposition-item') == []:
        try:
            driver.find_element_by_class_name('market-expansion-toggle').click()
            time.sleep(3)
        except:
            pass
        for j in driver.find_elements_by_class_name('proposition-wrapper'):
                whole = j.text.split('\n')
                Players.append(whole[0])
                Wins.append(whole[1])
                Top5.append('-')
                Top10.append('-')
                Top20.append('-')
                
    else:
        for j in driver.find_elements_by_class_name('proposition-item'):
            whole = j.text.split('\n')
            if len(whole) > 3:
                Players.append(whole[0])
                Wins.append(whole[1])
                Top5.append(whole[2])
                Top10.append(whole[3])
                Top20.append(whole[4])

            else:
                Players.append(whole[0])
                Wins.append(whole[1])
                Top5.append('-')
                Top10.append('-')
                Top20.append('-')

    for k in range(len(Players)):
        print(f"{Players[k]}  |  {Wins[k]}  |  {Top5[k]} | {Top10[k]}  |  {Top20[k]}")

    driver.back()
    time.sleep(3)
    print()
```
