```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import re
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get('https://www.topsport.com.au/')
driver.implicitly_wait(15)
driver.find_element_by_xpath("//li[@class=' homersmSport']").click()

for i in range (1, len(driver.find_elements_by_xpath(f"//div[@class='dynamicWidget']/dl/dt"))):
    if 'Golf' in driver.find_elements_by_xpath(f"//div[@class='dynamicWidget']/dl/dt")[i].text:
        lst = driver.find_elements_by_xpath(f"//div[@class='dynamicWidget']/dl/dd{[i-22]}/dl/dd/ul/li")
        
leagues = []
for i in range (1, len(lst)):
    txt = lst[i].get_attribute('onclick')
    league = re.findall(r'["](.*?)["]',txt)[0]
    leagues.append(league)

for j in leagues:
    try:
        driver.get('https://www.topsport.com.au/'+j)
        driver.implicitly_wait(15)

        league_name = driver.find_element_by_xpath("//h1[@class='icon icon-GOLF']").text
        print(f"____________________________________ {league_name} ________________________________________")
        print("")

        try:
            if 'no matches' in driver.find_element_by_xpath("//div[@class='ajaxReply warningText']").text:
                print("No match available for this Tournament !")
        except:
            for i in range(1, len(driver.find_elements_by_class_name("MatchMarket"))+1):
                try:
                    Team1 = driver.find_element_by_xpath(f"//div[@class='blank']/div[{i}]/div/table/tbody/tr/td").text
                    Team2 = driver.find_element_by_xpath(f"//div[@class='blank']/div[{i}]/div/table/tbody/tr[2]/td").text
                    Win1 = driver.find_element_by_xpath(f"//div[@class='blank']/div[{i}]/div/table/tbody/tr/td[2]/a/div").text
                    Win2 = driver.find_element_by_xpath(f"//div[@class='blank']/div[{i}]/div/table/tbody/tr[2]/td[2]/a/div").text
                    print(f"BET{i} ---> Player1: {Team1}, Win: {Win1}, Player2: {Team2}, Win: {Win2}")
                except:
                    try:
                        Team1 = driver.find_element_by_xpath(f"//div[@class='blank']/div[1]/div/table/tbody/tr/td").text
                        Team2 = driver.find_element_by_xpath(f"//div[@class='blank']/div[1]/div/table/tbody/tr[2]/td").text
                        print(f"BET{i} ---> Player1: {Team1}, Player2: {Team2}, No bets placed so far")
                    except:
                        print("The bets are ended")
        print("-------------------------------------------------------------------------------------------------------")
    except:
        continue
```
